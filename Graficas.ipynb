```python
list(axis)
```

```python
plt.plot(axis, [x**2 for x in axis])
```

```python
import math
axis = range(0, 80)
plt.plot(axis, [math.sin(x/10) for x in axis], 'x')
plt.title('La grafica de la funcion seno')
plt.ylabel('sin(x)')
plt.xlabel('x')
```
